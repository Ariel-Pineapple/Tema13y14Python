```python
class CuentaBancaria:
  def __init__(self, saldo):
    self.__saldo = saldo # atributo privado

    def depositar(self, cantidad):
      self.__saldo += cantidad

    def retirar(self):
      if cantidad > self.__saldo:
        print("Saldo insuficiente")
      else:
        self.__saldo -= cantidad

    def mostrar_saldo(self):
      print(f"Saldo actual: ${self.__saldo}")

    def obtener_saldo(self):
      return self.__saldo

    def establecer_saldo(self, nuevo_saldo):
          if nuevo_saldo >= 0:
            self.__saldo = nuevo_saldo
          else:
            print("El saldo no puede ser negativo")
```

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"{self.nombre} ({self.edad} años)"

    def guardar(self, archivo):
        # Guarda el objeto en el archivo usando 'a' (agregar)
        with open(archivo, 'a') as f:
            f.write(f"{self.nombre},{self.edad}\n")

    @staticmethod
    def cargar_todas(archivo):
        personas = []
        with open(archivo, 'r') as f:
            for linea in f:
                nombre, edad = linea.strip().split(',')
                personas.append(Persona(nombre, int(edad)))
        return personas

```

```python
# Crear varias personas.

personas = [
    Persona("Brenda Nalley", 26),
    Persona("Emilio Paez", 24),
    Persona("Mariana Amparo", 27),
    Persona("Saúl Rodriguez", 30),
    Persona("Gabriel Ramírez", 50),
    Persona("Daniel Ramírez", 50)
]

# Guardar el archivo

# Guardamos en el archivo
with open("personas.txt", "w") as f:
    pass  # Solo para crear o limpiar el archivo

for persona in personas:
    persona.guardar("personas.txt")

print("Personas guardadas en 'personas.txt'")

```

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def to_dict(self):
        return {"nombre": self.nombre, "edad": self.edad}

class Estudiante(Persona):
    def __init__(self, nombre, edad, matricula, carrera):
        super().__init__(nombre, edad)
        self.matricula = matricula
        self.carrera = carrera
        self.calificaciones = []

    def agregar_calificacion(self, materia, calificacion):
        self.calificaciones.append({"Materia:": materia, "Calificación": calificacion})
        print(f"Se agregó la calificación de {materia}: {calificacion} a {self.nombre}")

    def promedio(self):
      if not self.calificaciones:
        return 0
      total = sum(calificacion["Calificación"] for calificacion in self.calificaciones)
      return total / len(self.calificaciones)

    def to_dict(self):
        data = super().to_dict()
        data.update({
            "matricula": self.matricula,
            "carrera": self.carrera,
            "calificaciones": self.calificaciones
        })
        return data

    def from_dict(cls, data):
      e = cls(data["nombre"], data["edad"], data["matricula"], data["carrera"])
      e.calificaciones = data.get("calificaciones", [])
      return e

#
class Profesor(Persona):
    def __init__(self, nombre, edad, departamento):
        super().__init__(nombre, edad)
        self.departamento = departamento

    def calificar(self, estudiante, materia, calificacion):
        estudiante.agregar_calificacion(materia, calificacion)

    def to_dict(self):
        data = super().to_dict()
        data.update({
            "departamento": self.departamento
        })
        return data

    def from_dict(cls, data):
      return cls(data["nombre"], data["edad"], data["departamento"])
```

```python
from pickle import load
import json

class Universidad:
    def __init__(self, nombre, archivo_datos="universidad.json"):
        self.nombre = nombre
        self.estudiantes = []
        self.profesores = []
        self.archivo_datos = archivo_datos


    def agregue_estudiante(self, estudiante):
          self.estudiantes.append(estudiante)



    def agregar_profesor(self, profesor):
          self.profesores.append(profesor)

    def listar_profesores(self):
          print(f"Profesores de {self.nombre}:")
          for profesor in self.profesores:
              print(f"- {profesor.nombre} ({profesor.departamento})")


    def guardar_datos(self):
          data = {
              "nombre": self.nombre,
              "estudiantes": [e.to_dict() for e in self.estudiantes],
              "profesores": [p.to_dict() for p in self.profesores]
          }
          with open(self.archivo_datos, "w") as f:
              json.dump(data, f, indent=4)
          print("Datos guardados correctamente")


    def cargar_datos(self):
          try:
              with open(self.archivo_datos, "r") as f:
                  data = json.load(f)
              self.nombre = data["nombre"]
              self.estudiantes = [Estudiante.from_dict(e) for e in data["estudiantes"]]
              self.profesores = [Profesor.from_dict(p) for p in data["profesores"]]
              print("Datos cargados correctamente")
          except FileNotFoundError:
              print(f"El archivo {self.archivo_datos} no existe. Se creará uno nuevo.")
```
