### Solución: una función

Una función es una “receta” reutilizable.
- Le damos **un nombre**
- Puede recibir **datos de entrada** (parámetros)
- Puede **hacer algo** y opcionalmente **devolver** un resultado

Vamos a crear una función `saludar(nombre)` que genere el saludo.

```python
# EJEMPLO: Una función simple que saluda
def saludar(nombre):
  print(f"Hola: {nombre}")

saludar("Ariel")
```

```python
# Una posible solución sería:
# Crear una función que se llame saludar

def saludar(nombre):
    print(f"Hola {nombre}")


saludar("Ana")
saludar("Luis")
saludar("Maria")
saludar("Diana")
saludar("Alberto")

```

```python
# optimiza el código anterior trata de disminuir las líneas de código.

lista = ["Ana", "Luis", "Maria", "Diana", "Alberto"]

def saludar(nombre):
    print(f"Hola {nombre}")

for nombre in lista:
    saludar(nombre)

```

**Ahora tenemos:**

- Una función → reutiliza lógica
- Una lista → almacena datos
- Un for → automatiza la repetición

Esto ya es pensamiento computacional:
* Separar datos
* Separar lógica
* Automatizar procesos

## Parámetro vs Argumento

Estos dos términos suelen confundirse, pero no son lo mismo.

* Parámetro → variable que aparece en la definición de la función  
* Argumento → valor real que se envía al llamar la función

```python
def saludar(nombre):
    return f"Hola {nombre}"
```

En esta línea:

def saludar(nombre):

 "nombre" es un PARÁMETRO  
Es como una "caja vacía" que espera recibir un valor.

```python
saludar("Ana")
```

Resumen:

Definición → usamos PARÁMETROS  
Llamada → enviamos ARGUMENTOS  

Parámetro = variable interna  
Argumento = valor externo

# **Tema 14: Manejo de Archivos, Objetos y Clases**


**2.1 Clases y Objetos (POO)**


Explicación: Una clase es un molde (plano) y un objeto es la instancia creada a partir de ese molde. El método __init__ es el constructor.

```python
# EJEMPLO: Clase Estudiante
class Estudiante:
    def __init__(self, nombre, carrera):
        self.nombre = nombre
        self.carrera = carrera

    def presentarse(self):
        return f"Soy {self.nombre} y estudio {self.carrera}."

# Creamos un objeto
alumno1 = Estudiante("Ariel", "Diseño Industrial")
print(alumno1.presentarse())
```

**2.2 Manejo de Archivos (txt)**


Explicación: Usamos with open() para asegurar que el archivo se cierre correctamente después de usarlo. Los modos comunes son 'w' (escribir) y 'r' (leer).

```python
# EJEMPLO: Crear y escribir en un archivo
with open("notas.txt", "w") as archivo:
    archivo.write("Esta es la primera línea de mi archivo.\n")
    archivo.write("Python es genial para manejar datos.")

print("Archivo creado con éxito.")
```
